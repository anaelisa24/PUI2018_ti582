```python
# Backwards compatibility
from __future__ import print_function

# Import packages
import pylab as pl
import seaborn
from scipy.optimize import curve_fit, minimize

import numpy as np

%pylab inline
```

```python
# Set up a dictionary in order to hold the distribution data
myDist = {}
```

```python
# Set the size of the 100 distributions for values between 10 and 2000
np.random.seed(25)
distSize = np.random.randint(10,2000,100)

# Set the population mean
# This will be used for each of the 5 distributions that are being tested
myMean = 0.5
```

## Normal Distribution

```python
# Populate the dictionary with the normal distributions
np.random.seed(25)
myDist['normal'] = {}

for item in distSize:
    myDist['normal'][item] = np.random.normal(loc=myMean, size=item),

# Populate the dictionary with the means of the distributions and plot them
myDist['normal']['means'] = {}
axchisq_mean = pl.figure(figsize=(10,6)).add_subplot(111)

for x in myDist['normal']:
    if not type(x) == str:
        myDist['normal']['means'][x] = myDist['normal'][x][0].mean()
        axchisq_mean.plot(x, myDist['normal']['means'][x], 'o')

axchisq_mean.set_xlabel('Sample Size (n)', fontsize=18)
axchisq_mean.set_ylabel('Sample Mean (x-bar)', fontsize=18)
axchisq_mean.set_title('Normal Distribution', fontsize=18)
axchisq_mean.plot([min(distSize), max(distSize)],[0.5,0.5], 'k',label="Population Mean")
axchisq_mean.legend(fontsize=12)
```

**Figure 1**: The figure shows sample mean values of the Normal distribution for 100 different samples consisting of different sample sizes, randomly distributed between 10 and 2000. The graph shows that for larger sample sizes, the sample mean converges to the population mean, equal to 0.5 in this case. This matches the law of large numbers, which states that as sample size increases, the results should be closer to the expected value, which is 0.5 in this case.

```python
# Plot the histogram of the means
theMeans1 = list(myDist['normal']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(theMeans1,bins=30)
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('n', fontsize = 18)
pl.title('Normal Distribution', fontsize = 18)
```

**Figure 2**: The figure shows a histogram of the sample means for the Normal distribution. The shape of the histogram looks like a bell curve (a.k.a. the normal distribution). This matches the central limit theorem which states that the distribution of the means of a distribution approaches normal for many samples, even if the original distribution is not normally distributed.
