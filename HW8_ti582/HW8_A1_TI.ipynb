```python
# Import packages
import pandas as pd
import geopandas as gpd
import json
import urllib.request as urllib
import shutil
import requests
import io
import zipfile
import shapely
from fiona.crs import from_epsg
import numpy as np
import pylab as pl
%pylab inline
```

```python
# Check if PUIdata exists
import os
print("PUIdata", os.getenv("PUIDATA"))
```

```python
# Download the PUMA data
PUMAURL1 = 'https://data.cityofnewyork.us/api/geospatial/cwiz-gcty?method=export&format=Shapefile'
urllib.urlretrieve(PUMAURL1, "PUMAfile.gz")
!unzip -d $PUIDATA PUMAfile.gz
```

```python
# Convert to geodataframe
puma = gpd.GeoDataFrame.from_file(os.getenv("PUIDATA") + "/" + 
                                     "geo_export_059fdb10-dbfa-4a18-8b00-2bd4c9986d50.shp")
```

```python
# Convert the crs
puma = puma.to_crs(epsg = 2263)
```

```python
# Download the Vision Zero Seniors Centers data
VZurl1 = 'https://data.cityofnewyork.us/api/geospatial/xu8q-aip5?method=export&format=GeoJSON'
urllib.urlretrieve(VZurl1,"Seniors.geojson")
# Convert the data to a GeoDataFrame
SeniorCenters = gpd.GeoDataFrame.from_file('Seniors.geojson')
SeniorCenters.head()
```

```python
# Move the data to the PUIDATA folder
currentPath = os.getcwd()
newPath = os.getenv('PUIDATA')
shutil.move('%s/Seniors.geojson'%currentPath, '%s/Seniors.geojson'%newPath)
```

```python
# Download the Vision Zero Neighborhood Slow Zones data
VZurl2 = 'https://data.cityofnewyork.us/api/geospatial/y4nf-25nw?method=export&format=GeoJSON'
urllib.urlretrieve(VZurl2,"SlowZones.geojson")
# Convert the data to a GeoDataFrame
SlowZones = gpd.GeoDataFrame.from_file('SlowZones.geojson')
SlowZones.head()
```

```python
# Move the data to the PUIDATA folder
currentPath = os.getcwd()
newPath = os.getenv('PUIDATA')
shutil.move('%s/SlowZones.geojson'%currentPath, '%s/SlowZones.geojson'%newPath)
```

```python
# Convert the geodataframes to the NYC crs
SeniorCenters = SeniorCenters.to_crs(epsg = 2263)
SlowZones = SlowZones.to_crs(epsg = 2263)
```

```python
# Plot the Senior Centers and Neighborhood Slow Zones Data
Plot1 = puma.plot(figsize=(11,11),edgecolor='grey',facecolor='w')
SeniorCenters.plot(ax=Plot1,marker='.',markersize=30);
SlowZones.plot(ax=Plot1,edgecolor='black',alpha=0.55,facecolor='red');
Plot1.set_title('Senior Centers and Neighborhood Slow Zones in NYC',fontsize=20);
Plot1.set_axis_off()
```

**Figure 1**: The plot above shows locations of senior centers (blue dots) and neighborhood slow zones (red shaded areas) as provided by the Vision Zero team from NYC DOT. They are plotted on top of the NYC PUMA areas. There does not seem to be a correlation between the locations of the senior centers and the neighborhood slow zones.
