```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

# Task 2 - Working with date/time

```python
# Trips that started on Feb-02-2015 only
task2A = '''
SELECT * FROM citibike
WHERE starttime BETWEEN '2/2/2015 00:00:00.00' AND '2/2/2015 23:59:59.999'
ORDER BY starttime
'''
table2A = pd.read_csv(StringIO.StringIO(queryCartoDB(task2A)), sep=',')
table2A
```

```python
# Trips that started on the weekends
task2B = '''
SELECT * FROM citibike
WHERE starttime<='2/2/2015' OR starttime>='2/7/2015'
ORDER BY starttime
'''
table2B = pd.read_csv(StringIO.StringIO(queryCartoDB(task2B)), sep=',')
table2B
```

```python
# Average trip duration for trips during the weekends
task2C = '''
SELECT AVG(tripduration) FROM citibike
WHERE starttime<='2/2/2015' OR starttime>='2/7/2015'
'''
table2C = pd.read_csv(StringIO.StringIO(queryCartoDB(task2C)), sep=',')
table2C
```

```python
# Average trip duration for trips during the weekdays
task2C = '''
SELECT AVG(tripduration) FROM citibike
WHERE starttime>='2/2/2015' AND starttime<'2/7/2015'
'''
table2C = pd.read_csv(StringIO.StringIO(queryCartoDB(task2C)), sep=',')
table2C
```

# Task 3 - Working with Space

```python
# Show the list of start station locations
task3A = '''
SELECT start_station_name FROM citibike
GROUP BY start_station_name
'''
table3A = pd.read_csv(StringIO.StringIO(queryCartoDB(task3A)), sep=',')
table3A
```

```python
# Show the list of start station locations with the number of trips started per station 
# only for stations within 500m of Times Square
task3B = '''
SELECT start_station_name, COUNT(tripduration) FROM citibike
WHERE ACOS(SIN(RADIANS(start_station_latitude))*SIN(RADIANS(40.7577))
+COS(RADIANS(start_station_latitude))*COS(RADIANS(40.7577))*COS(RADIANS(start_station_longitude-(-73.9857))))*6371<=0.5
GROUP BY start_station_name
ORDER BY start_station_name ASC
'''
table3B = pd.read_csv(StringIO.StringIO(queryCartoDB(task3B)), sep=',')
table3B
```

# Task 4 - Putting it all together

```python
# Find the station with the longest average trip duration during weekends and within 500m of Times Square
task4A = '''
SELECT start_station_name, COUNT(tripduration), AVG(tripduration) FROM citibike
WHERE ACOS(SIN(RADIANS(start_station_latitude))*SIN(RADIANS(40.7577))
+COS(RADIANS(start_station_latitude))*COS(RADIANS(40.7577))*COS(RADIANS(start_station_longitude-(-73.9857))))*6371<=0.5
AND (starttime<='2/2/2015' OR starttime>='2/7/2015')
GROUP BY start_station_name
ORDER BY AVG(tripduration) DESC
LIMIT 1
'''
table4A = pd.read_csv(StringIO.StringIO(queryCartoDB(task4A)), sep=',')
table4A
```
